```python
import pandas as pd
```

```python
colunas = [
    'document_id', 'paciente_idade', 'paciente_dataNascimento', 'paciente_enumSexoBiologico', 'paciente_racaCor_valor',
    'estabelecimento_razaoSocial', 'estalecimento_noFantasia', 'estabelecimento_municipio_codigo', 
    'estabelecimento_municipio_nome', 'vacina_grupoAtendimento_codigo', 'vacina_grupoAtendimento_nome', 
    'vacina_categoria_codigo','vacina_categoria_nome', 'vacina_fabricante_nome', 
    'vacina_dataAplicacao','vacina_descricao_dose', 'vacina_codigo', 'vacina_nome'
]
```

```python
dados_vacinas = pd.read_csv('../DadosVacinação/DadosVacinaçãoRS.csv', sep=';', usecols=colunas)
```

```python
new_names = { 
             'paciente_idade': 'idade', 
             'paciente_dataNascimento': 'data_nascimento', 
             'paciente_enumSexoBiologico': 'sexo', 
             'paciente_racaCor_valor': 'raca',
             'estabelecimento_razaoSocial': 'razao_social', 
             'estalecimento_noFantasia': 'nome_fantasia', 
             'estabelecimento_municipio_codigo': 'codigo_municipio', 
             'estabelecimento_municipio_nome': 'nome_municipio',
             'vacina_grupoAtendimento_codigo': 'codigo_grupo', 
             'vacina_grupoAtendimento_nome': 'nome_grupo', 
             'vacina_categoria_codigo': 'codigo_categoria',
             'vacina_categoria_nome': 'nome_categoria', 
             'vacina_fabricante_nome': 'fabricante_vacina', 
             'vacina_dataAplicacao': 'data_aplicacao',
             'vacina_descricao_dose': 'dose', 
             'vacina_codigo': 'codigo_vacina', 
             'vacina_nome': 'nome_vacina'
            }

dados_vacinas = dados_vacinas.rename(columns=new_names)
```

```python
print(len(dados_vacinas))
dados_vacinas.head()
```

```python
print('Dados nulos:')
print(dados_vacinas.isna().sum())
```

```python
# Tratamento de colunas None, Null, NaN

dados_vacinas['sexo'] = dados_vacinas['sexo'].fillna('Não informado')
dados_vacinas['raca'] = dados_vacinas['raca'].fillna('Não informado')
dados_vacinas['nome_grupo'] = dados_vacinas['nome_grupo'].fillna('Desconhecido')
dados_vacinas['codigo_categoria'] = dados_vacinas['codigo_categoria'].fillna(99999)
dados_vacinas['nome_categoria'] = dados_vacinas['nome_categoria'].fillna('Desconhecida')
dados_vacinas['dose'] = dados_vacinas['dose'].fillna('Desconhecida')
dados_vacinas['nome_vacina'] = dados_vacinas['nome_vacina'].fillna('Desconhecida')
```

```python
print('Dados nulos:')
print(dados_vacinas.isna().sum())

# Idade será salva como Null e data_nascimento como NaT
```

```python
# Converter colunas datas para date
dados_vacinas['data_aplicacao'] = pd.to_datetime(dados_vacinas['data_aplicacao'], format='%Y-%m-%d', errors='coerce')
dados_vacinas['data_nascimento'] = pd.to_datetime(dados_vacinas['data_nascimento'], format='%Y-%m-%d', errors='coerce')
```

```python
# Função de limpeza e padronização
import unicodedata
import re

def limpeza(texto):
    if not isinstance(texto, str):
        return texto
    
    # remover espaços no início e fim
    texto = texto.strip()
    
    # converter para minúsculo
    texto = texto.lower()
    
    # remover acentos
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('ASCII')
    
    # remover caracteres especiais (mantém letras, números e espaço)
    texto = re.sub(r'[^a-z0-9\s]', '', texto)
    
    # remover múltiplos espaços
    texto = re.sub(r'\s+', ' ', texto)
    
    return texto
```

```python
for col in dados_vacinas.select_dtypes(include="object"):
    dados_vacinas[col] = dados_vacinas[col].apply(limpeza)
```

```python
print('Limpeza concluída')
dados_vacinas.head()
```

```python
# DIM PACIENTE

colunas_paciente = ['document_id', 'idade', 'data_nascimento', 'sexo', 'raca']

dim_paciente = dados_vacinas[colunas_paciente].drop_duplicates(subset='document_id').reset_index(drop=True)
dim_paciente['id_paciente'] = dim_paciente.index + 1

# Reorganizar as colunas
dim_paciente = dim_paciente[['id_paciente', 'document_id', 'idade', 'data_nascimento', 'sexo', 'raca']]

print('dim paciente criada com sucesso')
print('Pacientes: ', len(dim_paciente))

# Deve retornar TRUE isso garante que cada ID é único
print(dim_paciente['document_id'].is_unique)
dim_paciente.head()
```

```python
# DIM GRUPO_ATENDIMENTO

colunas_grupo_atendimento = ['nome_grupo', 'codigo_grupo']

dim_grupo_atendimento = (dados_vacinas[colunas_grupo_atendimento]
    .drop_duplicates(subset='codigo_grupo')
    .reset_index(drop=True))
dim_grupo_atendimento['id_grupo_atendimento'] = dim_grupo_atendimento.index + 1

# Reorganizar as colunas
dim_grupo_atendimento = dim_grupo_atendimento[['id_grupo_atendimento', 'codigo_grupo', 'nome_grupo']]

print('dim grupo_atendimento criada com sucesso')
print('Grupos de Atendimento: ', len(dim_grupo_atendimento))

# Deve retornar TRUE isso garante que cada ID é único
print(dim_grupo_atendimento['codigo_grupo'].is_unique)
dim_grupo_atendimento.head()
```

```python
# DIM CATEGORIA

colunas_categoria = ['codigo_categoria','nome_categoria']

dim_categoria = dados_vacinas[colunas_categoria].drop_duplicates(subset='codigo_categoria').reset_index(drop=True)
dim_categoria['id_categoria'] = dim_categoria.index + 1

# Reorganizar as colunas
dim_categoria = dim_categoria[['id_categoria','codigo_categoria','nome_categoria']]

print('dim categoria criada com sucesso')
print('Categorias: ', len(dim_categoria))

# Deve retornar TRUE isso garante que cada ID é único
print(dim_categoria['codigo_categoria'].is_unique)
dim_categoria.head()
```

```python
# DIM LOCALIDADE

colunas_localidade = ['codigo_municipio','nome_municipio']

dim_localidade = dados_vacinas[colunas_localidade].drop_duplicates(subset='codigo_municipio').reset_index(drop=True)
dim_localidade['id_localidade'] = dim_localidade.index + 1

# Reorganizar as colunas
dim_localidade = dim_localidade[['id_localidade','codigo_municipio','nome_municipio']]

print('dim localidade criada com sucesso')
print('Localidades: ', len(dim_categoria))

# Deve retornar TRUE isso garante que cada ID é único
print(dim_localidade['codigo_municipio'].is_unique)
dim_localidade.head()
```

```python
# DIM ESTABELECIMENTO

colunas_estabelecimento = ['razao_social','nome_fantasia']

dim_estabelecimento = dados_vacinas[colunas_estabelecimento].drop_duplicates().reset_index(drop=True)
dim_estabelecimento['id_estabelecimento'] = dim_estabelecimento.index + 1

# Reorganiza as colunas
dim_estabelecimento = dim_estabelecimento[['id_estabelecimento', 'razao_social','nome_fantasia']]

print('dim estabelecimento criada com sucesso')
print('Estabelecimentos: ', len(dim_estabelecimento))

# Deve retornar TRUE isso garante que cada ID é único
print(dim_estabelecimento['id_estabelecimento'].is_unique)
dim_estabelecimento.head()
```

```python
# DIM VACINA

colunas_vacina = ['nome_vacina', 'codigo_vacina', 'fabricante_vacina']

dim_vacina = dados_vacinas[colunas_vacina].drop_duplicates(subset='codigo_vacina').reset_index(drop=True)
dim_vacina['id_vacina'] = dim_vacina.index + 1

# Reorganiza as colunas
dim_vacina = dim_vacina[['id_vacina','fabricante_vacina','codigo_vacina','nome_vacina']]

print('dim vacina criada com sucesso')
print('Vacinas: ', len(dim_vacina))

# Deve retornar TRUE isso garante que cada ID é único
print(dim_vacina['codigo_vacina'].is_unique)
dim_vacina.head()
```

```python
# DIM DATA

# Data mínima e máxima dos atendimentos
data_inicio = '2020-01-01'
data_fim = dados_vacinas['data_aplicacao'].max()

# Sequência contínua de dias (sem buracos)
dias_continuos = pd.date_range(start=data_inicio, end=data_fim, freq='D')

# Transforme essa sequência no seu DataFrame de Dimensão
dim_data = pd.DataFrame({'data': dias_continuos})

# Criando o id_data no formato AAAAMMDD
dim_data['id_data'] = dim_data['data'].dt.strftime('%Y%m%d').astype(int)

# Os atributos da dimensão
dim_data['ano'] = dim_data['data'].dt.year
dim_data['mes'] = dim_data['data'].dt.month
dim_data['dia'] = dim_data['data'].dt.day
dim_data['dia_semana'] = dim_data['data'].dt.day_name()
dim_data['trimestre'] = dim_data['data'].dt.quarter

# Reorganizando as colunas
dim_data = dim_data[['id_data', 'data', 'ano', 'mes', 'dia', 'dia_semana', 'trimestre']]

print(dim_data.head())
```
